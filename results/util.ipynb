```python
import seaborn as sns
import numpy as np
sns.set(color_codes=True)
from matplotlib import pyplot as plt
```

```python
c1 = np.random.normal(0, 1, size=(100, 2))
c2 = np.random.normal(0, 1, size=(100, 2))
```

```python
sns.kdeplot(c1[:,0], c1[:, 1], shade=True, alpha=0.5, shade_lowest=False, label='Anonymized Client 1 Dataset')
sns.kdeplot(c2[:,0], c2[:, 1], shade=True, alpha=0.5, shade_lowest=False, label='Anonymized Client 2 Dataset')
sns.scatterplot([0],[0.5], marker='x' ,color='blue', label='Known Client 2 Data')
sns.scatterplot([-0.5],[-0.8], marker='x' ,color='red', label='Known Client 1 Data')
plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
c1 = np.random.normal([3,0], 1, size=(100, 2))
c2 = np.random.normal([-3,0], 1, size=(100, 2))
```

```python
sns.kdeplot(c1[:,0], c1[:, 1], shade=True, shade_lowest=False, label='Anonymized Client 1 Dataset')
sns.kdeplot(c2[:,0], c2[:, 1], shade=True, shade_lowest=False, label='Anonymized Client 2 Dataset')
sns.scatterplot([2],[0], marker='x' ,color='blue', label='Known Client 1 Point')
sns.scatterplot([-2],[-0.8], marker='x' ,color='red', label='Known Client 2 Point')
plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```
